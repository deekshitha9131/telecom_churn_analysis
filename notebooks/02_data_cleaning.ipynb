```python
import pandas as pd
df = pd.read_csv("../data/raw/Telco-Customer-Churn.csv")
```

```python
# Convert TotalCharges from object to numeric (blank strings become NaN)
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
df.isnull().sum()
df[df["TotalCharges"].isna()]
```

```python
# All 11 NaN rows have tenure=0 (new customers with no charges yet) — impute with 0
df["TotalCharges"] = df["TotalCharges"].fillna(0)
```

```python
df = df.drop("customerID", axis=1)
```

```python
print("Duplicates found:", df.duplicated().sum())
df = df.drop_duplicates()
print("Duplicates after drop:", df.duplicated().sum())
```

```python
# Final shape and dtype verification
df.info()
```

```python
df.to_csv("../data/processed/telco_cleaned.csv", index=False)
```
